```python
import pandas as pd
```

```python
### Loop the data lines
with open("/data/Annotation/dataset.txt", 'r') as data:
    # get No of columns in each line
    col_count = [ len(l.split(" ")) for l in data.readlines() ]

### Generate column names  (names will be 0, 1, 2, ..., maximum columns - 1)
column_names = [i for i in range(0, max(col_count))]

### Read csv
df = pd.read_csv("/data/Annotation/dataset.txt", header=None, delimiter=" ", names=column_names)
```

```python

list1=[]
list2=[]
p=df[0].tolist()
for index,name in enumerate(p):
    row = (df.loc[index,:]).tolist()
    cleanedrow = [x for x in row if str(x) != 'nan']
    if 'Rost' in name:
        list1.append(cleanedrow)
    else:
        list2.append(cleanedrow)
        
```

```python
list1
```

```python
f1 = open("/data/Annotation/list1.txt", 'w')
f2 = open("/data/Annotation/list2.txt", 'w')

for row in list1:
    string_row = ' '.join(row)
    print(string_row)
    f1.write(string_row+'\n')
    

for row in list2:
    string_row = ' '.join(row)
    print(string_row)
    f2.write(string_row+'\n')

    
    
    
```
